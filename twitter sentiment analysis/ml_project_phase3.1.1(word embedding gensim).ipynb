```python
import pandas as pd
import numpy as np
import ast
```

```python
import gensim
```

```python
data = pd.read_csv ("yarab.csv")
```

```python
data
```

```python
all_tweets = data["tweets"].copy()
```

```python
all_tweets
```

```python
ast.literal_eval(all_tweets[0])
```

```python
array_of_tweets = [];
```

```python
for i,tw in enumerate(all_tweets):
    array_of_tweets.append (gensim.models.doc2vec.TaggedDocument(ast.literal_eval(tw), [i]))
```

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=100, min_count=2, epochs=10)
```

```python
model.build_vocab(array_of_tweets)
```

```python
model.train(array_of_tweets, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vector = model.infer_vector(['i', 'love', 'cats'])
vector2 = model.infer_vector(['dog'])
```

```python
model.save("doc2vec.model")
```
